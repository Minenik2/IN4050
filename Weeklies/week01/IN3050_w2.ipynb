### 1. Given the function $ f(x) = -x^{4} + 2x^{3} + 2x^{2} - x $ and it's derivative $f'(x) = {- 4 x^3 + 6 x^2 + 4 x - 1}$

```python
# Standard import and functions
# Run this cell first
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -np.power(x, 4) + 2 * np.power(x, 3) + 2 * np.power(x, 2) - x

def df(x):
    return -4 * np.power(x, 3) + 6 * np.power(x, 2) + 4 * x - 1
```

#### a. Plot the function, and its gradient (derivative) from $x = -2$ to $x = 3$

```python
# Add your solution here
x_line = np.linspace(-2, 3)

plt.plot(x_line, f(x_line), label=r"$f(x) = -x^4 + 2x^3 + 2x^2 - x$")
plt.plot(x_line, df(x_line), label=r"$f'(x) = -4x^3 + 6x^2 + 4x - 1$")
plt.legend()
```

#### b. Maximize using gradient ascent. You can try step size 0.1 and start somewhere in the range $[-2, 3]$. 

Gradient ascent: $x_{(k+1)} = x_{(k)} + \gamma \nabla f(x_{(k)})$

```python
# Add your solution here
gamma = 0.05
iterations = 1000
x_start = np.random.uniform(-2, 3)

x = x_start

x_history = [x]
f_history = [f(x)]

for i in range(iterations):
    x = x + gamma * df(x)
    
    x_history.append(x)
    f_history.append(f(x))
    
plt.plot(x_line,f(x_line), label=r"$f(x) = -x^4 + 2x^3 + 2x^2 - x$", color="blue")
plt.scatter(x_history, f_history, zorder=5, color="red")
print(x_start)
## ask question about this
```
